```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat data yang sudah diproses
data = pd.read_csv('prepared_data.csv')
```

```python
# Bagian 1: Deskripsi Statistik
print("\nDeskripsi Statistik:")
print(data.describe())
```

```python
# Bagian 7: Visualisasi
# Visualisasi distribusi hasil diabetes
plt.figure(figsize=(12, 6))
sns.countplot(x='Outcome', data=data)
plt.title('Distribusi Hasil Diabetes')
plt.xlabel('Diabetes (0 = Tidak, 1 = Ya)')
plt.ylabel('Jumlah')
plt.show()

# Histogram untuk fitur Glucose
plt.figure(figsize=(12, 6))
sns.histplot(data['Glucose'], bins=30, kde=True)
plt.title('Distribusi Glukosa')
plt.xlabel('Glukosa')
plt.ylabel('Frekuensi')
plt.show()

# Boxplot untuk fitur BMI
plt.figure(figsize=(12, 6))
sns.boxplot(x='Outcome', y='BMI', data=data)
plt.title('Boxplot BMI berdasarkan Hasil Diabetes')
plt.xlabel('Diabetes (0 = Tidak, 1 = Ya)')
plt.ylabel('BMI')
plt.show()

# Pairplot untuk melihat hubungan antar fitur
sns.pairplot(data, hue='Outcome')
plt.title('Pairplot Fitur berdasarkan Hasil Diabetes')
plt.show()
```
